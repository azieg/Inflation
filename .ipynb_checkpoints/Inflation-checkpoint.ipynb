```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
plt.rcParams['font.size'] = 12.0
```

```python
def diffEq(z, t):
    '''Solving this differential equation returns values as [dψ/dτ, ψ, a]'''
    
    return np.array([-3.0*np.sqrt((8.0*π/3.0)
                                  *(.5*z[0]**2.0 + 3.0*z[1]**2.0
                                    /(8.0*π*ψₒ**2.0)))*z[0] 
                                     - 3.0*z[1]/(4.0*π*ψₒ**2.0), 
                     z[0], 
                     z[2]*np.sqrt(8.0*π/3.0
                                  *(.5*z[0]**2.0 + 3.0*z[1]**2.0
                                    /(8.0*π*ψₒ**2.0)))
                    ])

ψₒ = 1.0
dψₒ = 0.0
π = np.pi
scale_fac0 = 1.0

# Solve diffEq
init  = np.array([dψₒ, ψₒ, scale_fac0])
time  = np.arange(0, 100, .01)
ψsol   = odeint(diffEq, init, time)

# Calculate the energy density and pressure using values from the odeint solution
density  = .5*ψsol[..., 0]**2.0 + 3.0*ψsol[..., 1]**2.0/(8.0*π*ψₒ**2.0)
pressure = .5*ψsol[..., 0]**2.0 - 3.0*ψsol[..., 1]**2.0/(8.0*π*ψₒ**2.0)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(11, 8), dpi=100)

axs[0, 0].plot(time, ψsol[..., 1], color='blue')
axs[0, 0].set_title('Field Value')
axs[0, 0].set_xlabel('τ')
axs[0, 0].set_ylabel('ψ')
axs[0, 0].grid()

axs[0, 1].plot(time, ψsol[..., 2], color='red')
axs[0, 1].set_yscale('log')
axs[0, 1].set_title('Scale Factor')
axs[0, 1].set_xlabel('τ')
axs[0, 1].set_ylabel('log(a)')
axs[0, 1].grid()

axs[1, 0].plot(time, density, color='green')
axs[1, 0].set_title('Energy Density' )
axs[1, 0].set_xlabel('τ')
axs[1, 0].set_ylabel('ρ')
axs[1, 0].grid()

axs[1, 1].plot(time, pressure, color='black')
axs[1, 1].set_title('Pressure')
axs[1, 1].set_xlabel('τ')
axs[1, 1].set_ylabel('p')
axs[1, 1].grid()

fig.tight_layout()
#plt.savefig('inflaton_evolution', dpi=100)
plt.show()
```
